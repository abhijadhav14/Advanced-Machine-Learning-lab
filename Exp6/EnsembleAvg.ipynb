```python
# Import libraries
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import VotingRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import mean_squared_error
```

```python
# Load data
california = fetch_california_housing(as_frame=True)
X = california.data
y = california.target
```

```python
# Define models
lr = LinearRegression()
dt = DecisionTreeRegressor()
knn = KNeighborsRegressor()
```

```python
# Create average ensemble
avg = VotingRegressor(estimators=[('lr', lr), ('dt', dt), ('knn', knn)])
```

```python
# Fit ensemble on data
avg.fit(X, y)
```

```python
# Make predictions
y_pred = avg.predict(X)
```

```python
# Evaluate performance
mse = mean_squared_error(y, y_pred)
print(f'MSE: {mse:.2f}')
```

```python
import matplotlib.pyplot as plt
```

```python

model_names = ['Linear Regression', 'Decision Tree', 'KNN']
accuracy_scores = [0.85, 0.92, 0.88]

plt.figure(figsize=(8, 6))
plt.bar(model_names, accuracy_scores, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Accuracy Comparison of Different Models")
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--')


for i, v in enumerate(accuracy_scores):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center', va='bottom')

plt.show()
```
